# Бета-распределение
$f_X(x) = \frac{1}{\mathrm{B}(a, b)}\, x^{a - 1} (1-x)^{b - 1}$

$\mathrm{B}(a,b) = \int\limits_0^1 x^{a - 1} (1-x)^{b - 1}\, dx$ - бета функция
### Генерация выборки, построение гистограммы и теоретической плотности

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
%matplotlib inline

#генерация выборки
a = 0.5
b = 0.5
sample = beta.rvs(a, b, size = 1000)

#построение гистограммы выборки
plt.hist(sample, bins = 40, normed = True, label = 'hist')
plt.ylabel('$f(x)$')
plt.xlabel('x')
plt.title('Beta r.v.')

#построение теоретической плотности распределения
x = np.linspace(0, 1, 1000)
pdf = beta.pdf(x, a, b)
plt.plot(x, pdf, 'r^', label = 'theoretical pdf')
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.)
```

### Генерация выборок размера n=5 и аппроксимация гистограммы средних нормальным распределением

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import beta
from scipy.stats import norm
%matplotlib inline

#генерация выборок
a = 0.5
b = 0.5
mean = np.empty(1000)
for i in range (0, 1000):
    sample = beta.rvs(a, b, size = 5)
    mean_part = np.sum(sample) / 5
    mean[i] = mean_part
    i = i + 1

#построение гистограммы средних
plt.hist(mean, bins = 40, normed = True, label = 'Mean hist')
plt.ylabel('Number of samples')
plt.xlabel('Mean')
plt.title('Mean hist and Normal PDF for n=5')

#вычисление параметров аппроксимирующего нормального распределения
mean = beta.mean(a,b)
var = beta.var(a,b)
mu = mean
sigma = np.sqrt(var / 5)

#построение аппроксимирующео нормального распределения
x = np.linspace(0, 1, 1000)
plt.plot(x, norm(loc = mu, scale = sigma).pdf(x), 'r^', label = 'Normal approximation')
plt.legend(bbox_to_anchor = (1.05, 1), loc = 2, borderaxespad = 0.)
```
